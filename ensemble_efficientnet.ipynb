```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Dataset
import torchvision.models as models
from torchvision.models import (
    efficientnet_b5, EfficientNet_B5_Weights,
    efficientnet_b4, EfficientNet_B4_Weights,
    efficientnet_b3, EfficientNet_B3_Weights
)

import albumentations as A
from albumentations.pytorch import ToTensorV2

# CONFIG
CONFIG = dict(
    epochs=10,
    max_lr=3e-4,
    batch_size=32,
    weight_decay=0.0002,
    hid_dim=512,
    dropout=0.5,
    num_workers=os.cpu_count(),
    # Ensemble specific configs
    num_models=3,  # Number of models in ensemble
    ensemble_method='weighted_average',  # 'simple_average', 'weighted_average', 'stacking'
    k_folds=5  # For cross-validation ensemble
)

# Custom Dataset using Albumentations
class ImageDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.classes = sorted(os.listdir(root_dir))
        self.image_paths = []
        self.labels = []
        self.transform = transform

        for idx, label in enumerate(self.classes):
            label_dir = os.path.join(root_dir, label)
            for file in os.listdir(label_dir):
                if file.endswith(('jpg', 'png', 'jpeg')):
                    self.image_paths.append(os.path.join(label_dir, file))
                    self.labels.append(idx)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image = cv2.imread(self.image_paths[idx])
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        if self.transform:
            image = self.transform(image=image)['image']

        label = self.labels[idx]
        return image, label

# Multiple Transform Variations for Ensemble Diversity
def get_train_transforms(variant=0):
    transforms = [
        # Variant 0: Standard augmentation
        A.Compose([
            A.RandomResizedCrop(height=224, width=224, scale=(0.8, 1.0), p=1.0),
            A.Rotate(limit=30, p=0.5),
            A.HorizontalFlip(p=0.5),
            A.VerticalFlip(p=0.2),
            A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
            ToTensorV2()
        ]),
        # Variant 1: Aggressive augmentation
        A.Compose([
            A.RandomResizedCrop(height=224, width=224, scale=(0.7, 1.0), p=1.0),
            A.Rotate(limit=45, p=0.7),
            A.HorizontalFlip(p=0.5),
            A.VerticalFlip(p=0.3),
            A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1, p=0.5),
            A.GaussNoise(var_limit=(10.0, 50.0), p=0.3),
            A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
            ToTensorV2()
        ]),
        # Variant 2: Conservative augmentation
        A.Compose([
            A.RandomResizedCrop(height=224, width=224, scale=(0.9, 1.0), p=1.0),
            A.Rotate(limit=15, p=0.3),
            A.HorizontalFlip(p=0.5),
            A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
            ToTensorV2()
        ])
    ]
    return transforms[variant % len(transforms)]

val_transform = A.Compose([
    A.Resize(height=224, width=224),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2()
])

# Load data with different transforms for ensemble
def load_data_ensemble(data_dir, transform_variant=0):
    train_transform = get_train_transforms(transform_variant)
    full_dataset = ImageDataset(data_dir, transform=train_transform)
    val_size = int(0.2 * len(full_dataset))
    train_size = len(full_dataset) - val_size
    train_ds, val_ds = random_split(full_dataset, [train_size, val_size])
    val_ds.dataset.transform = val_transform

    train_loader = DataLoader(train_ds, batch_size=CONFIG['batch_size'], shuffle=True, 
                              num_workers=CONFIG['num_workers'], persistent_workers=True)
    val_loader = DataLoader(val_ds, batch_size=CONFIG['batch_size'], shuffle=False, 
                            num_workers=CONFIG['num_workers'], persistent_workers=True)

    return train_loader, val_loader, len(full_dataset.classes)

# Multiple Model Architectures for Ensemble
class EfficientNetB5Classifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        weights = EfficientNet_B5_Weights.DEFAULT
        self.backbone = efficientnet_b5(weights=weights)
        in_features = self.backbone.classifier[1].in_features
        self.backbone.classifier = nn.Sequential(
            nn.Linear(in_features, 500),
            nn.ReLU(),
            nn.Dropout(CONFIG['dropout']),
            nn.Linear(500, num_classes)
        )

    def forward(self, x):
        return self.backbone(x)

class EfficientNetB4Classifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        weights = EfficientNet_B4_Weights.DEFAULT
        self.backbone = efficientnet_b4(weights=weights)
        in_features = self.backbone.classifier[1].in_features
        self.backbone.classifier = nn.Sequential(
            nn.Linear(in_features, 400),
            nn.ReLU(),
            nn.Dropout(CONFIG['dropout']),
            nn.Linear(400, num_classes)
        )

    def forward(self, x):
        return self.backbone(x)

class EfficientNetB3Classifier(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        weights = EfficientNet_B3_Weights.DEFAULT
        self.backbone = efficientnet_b3(weights=weights)
        in_features = self.backbone.classifier[1].in_features
        self.backbone.classifier = nn.Sequential(
            nn.Linear(in_features, 300),
            nn.ReLU(),
            nn.Dropout(CONFIG['dropout']),
            nn.Linear(300, num_classes)
        )

    def forward(self, x):
        return self.backbone(x)

# Ensemble Class
class EnsembleClassifier(nn.Module):
    def __init__(self, models, weights=None, method='weighted_average'):
        super().__init__()
        self.models = nn.ModuleList(models)
        self.method = method
        if weights is None:
            self.weights = [1.0 / len(models)] * len(models)
        else:
            self.weights = weights
        
        # For stacking method
        if method == 'stacking':
            self.meta_classifier = nn.Sequential(
                nn.Linear(len(models), 64),
                nn.ReLU(),
                nn.Dropout(0.3),
                nn.Linear(64, 1)
            )

    def forward(self, x):
        if self.method == 'simple_average':
            outputs = []
            for model in self.models:
                outputs.append(torch.softmax(model(x), dim=1))
            return torch.mean(torch.stack(outputs), dim=0)
        
        elif self.method == 'weighted_average':
            outputs = []
            for i, model in enumerate(self.models):
                weighted_output = torch.softmax(model(x), dim=1) * self.weights[i]
                outputs.append(weighted_output)
            return torch.sum(torch.stack(outputs), dim=0)
        
        elif self.method == 'stacking':
            # Get predictions from all models
            model_outputs = []
            for model in self.models:
                pred = torch.max(torch.softmax(model(x), dim=1), dim=1)[0]
                model_outputs.append(pred.unsqueeze(1))
            
            # Stack predictions and pass through meta-classifier
            stacked = torch.cat(model_outputs, dim=1)
            meta_pred = self.meta_classifier(stacked)
            
            # For now, return weighted average (can be enhanced)
            outputs = []
            for i, model in enumerate(self.models):
                outputs.append(torch.softmax(model(x), dim=1) * self.weights[i])
            return torch.sum(torch.stack(outputs), dim=0)

    def predict_individual(self, x):
        """Get predictions from individual models"""
        predictions = []
        for model in self.models:
            with torch.no_grad():
                pred = torch.softmax(model(x), dim=1)
                predictions.append(pred)
        return predictions

# Training Function with Mixed Precision and Checkpointing
from torch.cuda.amp import autocast, GradScaler

def save_checkpoint(model, optimizer, scheduler, scaler, epoch, best_val_acc, model_name):
    """Save training checkpoint"""
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'scheduler_state_dict': scheduler.state_dict(),
        'scaler_state_dict': scaler.state_dict(),
        'best_val_acc': best_val_acc,
        'config': CONFIG
    }
    checkpoint_path = f"checkpoint_{model_name}.pth"
    torch.save(checkpoint, checkpoint_path)
    print(f"💾 Checkpoint saved: {checkpoint_path}")
    return checkpoint_path

def load_checkpoint(model, optimizer, scheduler, scaler, model_name, device):
    """Load training checkpoint if exists"""
    checkpoint_path = f"checkpoint_{model_name}.pth"
    
    if os.path.exists(checkpoint_path):
        print(f"🔄 Resuming from checkpoint: {checkpoint_path}")
        checkpoint = torch.load(checkpoint_path, map_location=device)
        
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        scheduler.load_state_dict(checkpoint['scheduler_state_dict'])
        scaler.load_state_dict(checkpoint['scaler_state_dict'])
        
        start_epoch = checkpoint['epoch'] + 1
        best_val_acc = checkpoint['best_val_acc']
        
        print(f"✅ Resumed from epoch {start_epoch}, best val acc: {best_val_acc*100:.2f}%")
        return start_epoch, best_val_acc, True
    else:
        print(f"🆕 No checkpoint found, starting fresh training for {model_name}")
        return 0, 0.0, False

def train_individual_model(model, train_loader, val_loader, criterion, optimizer, scheduler, device, model_name):
    model.to(device)
    scaler = GradScaler()
    best_model_path = f"best_{model_name}.pth"
    
    # Try to load checkpoint
    start_epoch, best_val_acc, resumed = load_checkpoint(model, optimizer, scheduler, scaler, model_name, device)
    
    print(f"\n🚀 Training {model_name}...")
    if resumed:
        print(f"🔄 Resuming from epoch {start_epoch + 1}/{CONFIG['epochs']}")
    
    for epoch in range(start_epoch, CONFIG['epochs']):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for imgs, labels in tqdm(train_loader, desc=f"{model_name} - Epoch {epoch+1}/{CONFIG['epochs']}"):
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad()

            with autocast():
                outputs = model(imgs)
                loss = criterion(outputs, labels)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            scheduler.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        val_acc, _, _, _ = evaluate_model(model, val_loader, device)
        print(f"{model_name} - Train Loss: {running_loss/len(train_loader):.4f}, Train Acc: {100*correct/total:.2f}%, Val Acc: {100*val_acc:.2f}%")

        # Save best model
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), best_model_path)
            print(f"✅ Saved Best {model_name} with Val Acc: {100*best_val_acc:.2f}%")

        # Save checkpoint every epoch
        save_checkpoint(model, optimizer, scheduler, scaler, epoch, best_val_acc, model_name)

    # Clean up checkpoint after successful completion
    checkpoint_path = f"checkpoint_{model_name}.pth"
    if os.path.exists(checkpoint_path):
        os.remove(checkpoint_path)
        print(f"🧹 Cleaned up checkpoint: {checkpoint_path}")

    return best_val_acc, best_model_path

# Progress Tracking
def save_ensemble_progress(trained_models, model_accuracies, model_paths, completed_models, num_classes):
    """Save ensemble training progress"""
    progress_data = {
        'completed_models': completed_models,
        'model_accuracies': model_accuracies,
        'model_paths': model_paths,
        'num_classes': num_classes,
        'config': CONFIG
    }
    torch.save(progress_data, "ensemble_progress.pth")
    print(f"💾 Ensemble progress saved: {completed_models} models completed")

def load_ensemble_progress():
    """Load ensemble training progress if exists"""
    progress_path = "ensemble_progress.pth"
    if os.path.exists(progress_path):
        print(f"🔄 Loading ensemble progress...")
        progress_data = torch.load(progress_path)
        return (
            progress_data['completed_models'],
            progress_data['model_accuracies'],
            progress_data['model_paths'],
            progress_data['num_classes'],
            True
        )
    return 0, [], [], None, False

# Ensemble Training with Resume Support
def train_ensemble(data_dir, device):
    """Train multiple models with different configurations"""
    model_configs = [
        {'class': EfficientNetB5Classifier, 'name': 'EfficientNet-B5', 'transform_variant': 0},
        {'class': EfficientNetB4Classifier, 'name': 'EfficientNet-B4', 'transform_variant': 1},
        {'class': EfficientNetB3Classifier, 'name': 'EfficientNet-B3', 'transform_variant': 2}
    ]
    
    # Try to load previous progress
    completed_models, model_accuracies, model_paths, num_classes, resumed = load_ensemble_progress()
    
    if resumed:
        print(f"🔄 Resuming ensemble training from model {completed_models + 1}")
        # Get number of classes from saved progress
        temp_loader, _, _ = load_data_ensemble(data_dir, 0)
        if num_classes is None:
            num_classes = len(temp_loader.dataset.classes)
    else:
        # Get number of classes
        temp_loader, _, num_classes = load_data_ensemble(data_dir, 0)
        completed_models = 0
        model_accuracies = []
        model_paths = []
    
    trained_models = []
    
    # Load already completed models
    for i in range(completed_models):
        config = model_configs[i]
        model = config['class'](num_classes=num_classes)
        if os.path.exists(model_paths[i]):
            model.load_state_dict(torch.load(model_paths[i]))
            print(f"✅ Loaded completed model {i+1}: {config['name']} (Acc: {model_accuracies[i]*100:.2f}%)")
        trained_models.append(model)
    
    # Continue training from where we left off
    for i in range(completed_models, CONFIG['num_models']):
        config = model_configs[i]
        print(f"\n{'='*50}")
        print(f"Training Model {i+1}/{CONFIG['num_models']}: {config['name']}")
        print(f"{'='*50}")
        
        # Load data with specific transform variant
        train_loader, val_loader, _ = load_data_ensemble(data_dir, config['transform_variant'])
        
        # Initialize model
        model = config['class'](num_classes=num_classes)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=CONFIG['max_lr'], weight_decay=CONFIG['weight_decay'])
        scheduler = torch.optim.lr_scheduler.OneCycleLR(
            optimizer,
            max_lr=CONFIG['max_lr'],
            steps_per_epoch=len(train_loader),
            epochs=CONFIG['epochs']
        )
        
        # Train model
        val_acc, model_path = train_individual_model(
            model, train_loader, val_loader, criterion, optimizer, scheduler, device, config['name']
        )
        
        trained_models.append(model)
        model_accuracies.append(val_acc)
        model_paths.append(model_path)
        
        # Save progress after each model completion
        save_ensemble_progress(trained_models, model_accuracies, model_paths, i + 1, num_classes)
    
    # Clean up progress file after successful completion
    progress_path = "ensemble_progress.pth"
    if os.path.exists(progress_path):
        os.remove(progress_path)
        print(f"🧹 Cleaned up ensemble progress file")
    
    return trained_models, model_accuracies, model_paths, num_classes

# Manual Resume Helper Function
def resume_specific_model(data_dir, device, model_name, model_index):
    """Manually resume training for a specific model"""
    model_configs = [
        {'class': EfficientNetB5Classifier, 'name': 'EfficientNet-B5', 'transform_variant': 0},
        {'class': EfficientNetB4Classifier, 'name': 'EfficientNet-B4', 'transform_variant': 1},
        {'class': EfficientNetB3Classifier, 'name': 'EfficientNet-B3', 'transform_variant': 2}
    ]
    
    # Get number of classes
    temp_loader, _, num_classes = load_data_ensemble(data_dir, 0)
    
    config = model_configs[model_index]
    print(f"\n{'='*50}")
    print(f"Manually Resuming Model: {config['name']}")
    print(f"{'='*50}")
    
    # Load data with specific transform variant
    train_loader, val_loader, _ = load_data_ensemble(data_dir, config['transform_variant'])
    
    # Initialize model
    model = config['class'](num_classes=num_classes)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=CONFIG['max_lr'], weight_decay=CONFIG['weight_decay'])
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=CONFIG['max_lr'],
        steps_per_epoch=len(train_loader),
        epochs=CONFIG['epochs']
    )
    
    # Train model (will automatically resume from checkpoint if available)
    val_acc, model_path = train_individual_model(
        model, train_loader, val_loader, criterion, optimizer, scheduler, device, config['name']
    )
    
    print(f"✅ Model {config['name']} completed with validation accuracy: {val_acc*100:.2f}%")
    return val_acc, model_path
def calculate_ensemble_weights(accuracies, method='performance_based'):
    if method == 'equal':
        return [1.0 / len(accuracies)] * len(accuracies)
    elif method == 'performance_based':
        # Weight based on validation accuracy
        total_acc = sum(accuracies)
        return [acc / total_acc for acc in accuracies]
    elif method == 'softmax':
        # Softmax weighting
        exp_accs = [np.exp(acc * 10) for acc in accuracies]  # Scale for softmax
        sum_exp = sum(exp_accs)
        return [exp_acc / sum_exp for exp_acc in exp_accs]

# Evaluation
def evaluate_model(model, loader, device):
    model.eval()
    correct = 0
    total = 0
    y_true = []
    y_pred = []
    class_correct = {}
    class_total = {}

    with torch.no_grad():
        for imgs, labels in loader:
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            y_true.extend(labels.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

            for label, prediction in zip(labels.cpu().numpy(), predicted.cpu().numpy()):
                class_total[label] = class_total.get(label, 0) + 1
                if label == prediction:
                    class_correct[label] = class_correct.get(label, 0) + 1

    acc = correct / total
    label_wise_accuracy = {
        label: class_correct.get(label, 0) / class_total[label]
        for label in class_total
    }

    return acc, y_true, y_pred, label_wise_accuracy

# Ensemble Evaluation
def evaluate_ensemble(ensemble_model, individual_models, loader, device, class_names):
    print("\n" + "="*60)
    print("ENSEMBLE EVALUATION RESULTS")
    print("="*60)
    
    # Evaluate individual models
    individual_accuracies = []
    for i, model in enumerate(individual_models):
        acc, _, _, _ = evaluate_model(model, loader, device)
        individual_accuracies.append(acc)
        print(f"Individual Model {i+1} Accuracy: {acc*100:.2f}%")
    
    # Evaluate ensemble
    ensemble_acc, y_true, y_pred, label_wise_acc = evaluate_model(ensemble_model, loader, device)
    print(f"\n🎯 Ensemble Accuracy: {ensemble_acc*100:.2f}%")
    print(f"📈 Improvement over best individual: {(ensemble_acc - max(individual_accuracies))*100:.2f}%")
    
    return ensemble_acc, y_true, y_pred, label_wise_acc

# Label-wise Accuracy Printer
def print_label_wise_accuracy(label_wise_accuracy, class_names):
    print("\n📊 Label-wise Accuracy:")
    for idx, acc in label_wise_accuracy.items():
        print(f"{class_names[idx]}: {acc*100:.2f}%")

# Confusion Matrix
def plot_confusion_matrix(y_true, y_pred, class_names, title="Confusion Matrix"):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Save Ensemble
def save_ensemble(ensemble_model, model_paths, ensemble_weights, save_path="ensemble_model.pth"):
    ensemble_data = {
        'model_paths': model_paths,
        'weights': ensemble_weights,
        'method': CONFIG['ensemble_method'],
        'num_models': CONFIG['num_models']
    }
    torch.save(ensemble_data, save_path)
    print(f"💾 Ensemble configuration saved to {save_path}")

# Load Ensemble
def load_ensemble(save_path, num_classes, device):
    ensemble_data = torch.load(save_path)
    
    # Recreate models
    model_classes = [EfficientNetB5Classifier, EfficientNetB4Classifier, EfficientNetB3Classifier]
    models = []
    
    for i, model_path in enumerate(ensemble_data['model_paths']):
        model = model_classes[i](num_classes)
        model.load_state_dict(torch.load(model_path))
        model.to(device)
        models.append(model)
    
    # Create ensemble
    ensemble = EnsembleClassifier(
        models, 
        weights=ensemble_data['weights'], 
        method=ensemble_data['method']
    )
    ensemble.to(device)
    
    return ensemble, models

# Main
if __name__ == '__main__':
    DATA_DIR = '/home/prateek/code/CD_Projects/Classifier/dataset'
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    
    print(f"🔥 Starting Ensemble Training with {CONFIG['num_models']} models")
    print(f"📱 Device: {device}")
    print(f"🎯 Ensemble Method: {CONFIG['ensemble_method']}")
    
    # Train ensemble
    trained_models, model_accuracies, model_paths, num_classes = train_ensemble(DATA_DIR, device)
    
    # Calculate ensemble weights
    ensemble_weights = calculate_ensemble_weights(model_accuracies, method='performance_based')
    print(f"\n⚖️ Ensemble Weights: {[f'{w:.3f}' for w in ensemble_weights]}")
    
    # Create ensemble
    ensemble_model = EnsembleClassifier(
        trained_models, 
        weights=ensemble_weights, 
        method=CONFIG['ensemble_method']
    )
    ensemble_model.to(device)
    
    # Load validation data for evaluation
    _, val_loader, _ = load_data_ensemble(DATA_DIR, 0)
    class_names = val_loader.dataset.dataset.classes
    
    # Evaluate ensemble
    ensemble_acc, y_true, y_pred, label_wise_acc = evaluate_ensemble(
        ensemble_model, trained_models, val_loader, device, class_names
    )
    
    # Plot results
    plot_confusion_matrix(y_true, y_pred, class_names, "Ensemble Confusion Matrix")
    print_label_wise_accuracy(label_wise_acc, class_names)
    
    # Save ensemble
    save_ensemble(ensemble_model, model_paths, ensemble_weights)
    
    print(f"\n🎉 Ensemble training completed!")
    print(f"🏆 Final Ensemble Accuracy: {ensemble_acc*100:.2f}%")
```

```python

from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

def plot_calibration_curve(y_true, y_probs, n_bins=10):
    prob_true, prob_pred = calibration_curve(y_true, y_probs, n_bins=n_bins, strategy='uniform')
    plt.figure(figsize=(8, 6))
    plt.plot(prob_pred, prob_true, marker='o', label='Model Calibration')
    plt.plot([0, 1], [0, 1], linestyle='--', label='Perfectly Calibrated')
    plt.xlabel('Confidence (Predicted Probability)')
    plt.ylabel('Accuracy')
    plt.title('Reliability Diagram (Calibration Curve)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

```
